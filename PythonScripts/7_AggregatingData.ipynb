Aggregating data using transform(): 

```python
# Creating a new column named 'max_order' which is filled with maximum orders made by each user.
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Removing default Jupyter display max so we can see top 100 rows.
pd.options.display.max_rows = None
```

```python
ords_prods_merge.head(100)
```

Creating new column named 'loyalty_flag' categorizing customers into 'loyal', 'regular', and 'new':

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

Printing parts of the dataframe to only see 'user_id', 'loyalty_flag', and 'order_number' columns:

```python
ords_prods_merge[['user_id','loyalty_flag', 'order_number']].head(60)
```

Aggregating the average orders made by 'department_id':

```python
ords_prods_merge.groupby('department_id').agg({'order_number': ['mean']})
```

We want to investigate whether there's a difference between spending habits of the three types of customers, using the loyalty flag:

```python
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

Loyal customers tend to purchase items with mean price of 10.39, whereas New customers tend to purchase items with mean price of 13.29 and Regular customers with price of 12.49. Loyal customers make more frequent orders with lower priced items than New and Regular customers. 

Creating a new column named 'spender_type_flag' based on the average price of items purchased by user. The criteria is as follows:
 If the mean of the prices of products purchased by a user < 10, flag as "Low spender"
 If the mean of the prices of products purchased by a user >= 10, flas as "High spender"

```python
# Creating a new column 'mean_price_of_product' with value of mean of prices of products purchased by user
ords_prods_merge['mean_price_of_product'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Displaying minimized dataframe of columns to check if aggregation is accurate. 
ords_prods_merge[['user_id', 'prices', 'mean_price_of_product']].head(100)
```

```python
# Creating flag as "Low spender" or "High spender" using loc() function.
ords_prods_merge.loc[ords_prods_merge['mean_price_of_product'] < 10, 'spender_type_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_price_of_product'] >= 10, 'spender_type_flag'] = 'High spender'
```

```python
ords_prods_merge['spender_type_flag'].value_counts(dropna = False)
```

From this value count, we see that there are 634,245 spenders that are "High spenders".

Creating a new column named 'customer_frequency_flag' which categorizes the regularity of a user's ordering behavior according to the median in the "days_since_prior_order" column. Criteria is as follows: 
 If median of "days_since_prior_order" is > 20, flag as "Non-frequent customer"
 If median of "days_since_prior_order" is > 10 & <= 20, flag as "Regular customer"
 If median of "days_since_prior_order" is <= 10, flag as "Frequent customer"

```python
ords_prods_merge['median_days_since_prior_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] > 20, 'customer_frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days_since_prior_order'] > 10) & (ords_prods_merge['median_days_since_prior_order'] <= 20), 'customer_frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] <= 10, 'customer_frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge['customer_frequency_flag'].value_counts(dropna = False)
```

This analysis shows us that the majority of the Instacart customers are categorically "Frequent customers". In addition, we can also see that there were 5 users that made a single, first-time order, and never made an order again (see the row "NaN").
